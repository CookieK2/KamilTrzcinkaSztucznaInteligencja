```python
#zad 1: jednowymiarowa macierz
import numpy as np

def multiply_1d_matrices(matrix1, matrix2):
    result = np.dot(matrix1, matrix2)
    return result
matrix1 = np.array([1, 3, 5])
matrix2 = np.array([11, 9, 7])
result = multiply_1d_matrices(matrix1, matrix2)
print(result)

matrix1 = np.array([2, 4, 6])
matrix2 = np.array([1, 2, 3])
result = multiply_1d_matrices(matrix1, matrix2)
print(result)

matrix1 = np.array([22, 33, 44])
matrix2 = np.array([25, 14, 9])
result = multiply_1d_matrices(matrix1, matrix2)
print(result)
```

```python
#zad 1: dwuwymiarowa macierz
import numpy as np

def multiply_2d_matrices(matrix1, matrix2):
    result = np.matmul(matrix1, matrix2)
    return result
matrix1 = np.array([[1, 3], [5, 7]])
matrix2 = np.array([[2, 4], [6, 8]])
result = multiply_2d_matrices(matrix1, matrix2)
print(result)

matrix1 = np.array([[3, 2], [2, 5]])
matrix2 = np.array([[6, 7], [4, 9]])
result = multiply_2d_matrices(matrix1, matrix2)
print(result)

matrix1 = np.array([[2, 6], [4, 2]])
matrix2 = np.array([[1, 2], [8, 9]])
result = multiply_2d_matrices(matrix1, matrix2)
print(result)
```

```python
#zad 1: wielowymiarowa macierz
import numpy as np

def multiply_nd_matrices(matrix1, matrix2):
    result = np.tensordot(matrix1, matrix2, axes=(-1, 0))
    return result
matrix1 = np.array([[[4, 2], [7, 3]], [[4, 7], [3, 9]]])
matrix2 = np.array([[1, 2], [5, 4]])
result = multiply_nd_matrices(matrix1, matrix2)
print(result)

matrix1 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
matrix2 = np.array([[15, 11], [10, 9]])
result = multiply_nd_matrices(matrix1, matrix2)
print(result)

matrix1 = np.array([[[1, 3], [6, 8]], [[1, 4], [12, 22]]])
matrix2 = np.array([[4, 7], [9, 3]])
result = multiply_nd_matrices(matrix1, matrix2)
print(result)
```

```python
#zad 2: 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = ['Ziemniaki', 'Ogóreki', 'Pomidory', 'Kurczak']
values = [25, 12, 7, 50]
data = pd.DataFrame({'Dane': data, 'Wartość': values})

plt.figure(figsize=(5, 5))
plt.pie(data['Wartość'], labels=data['Dane'], autopct='%1.1f%%')
plt.title('Wykres kołowy')
plt.show()

plt.figure(figsize=(8, 4))
plt.bar(data['Dane'], data['Wartość'])
plt.title('Wykres słupkowy')
plt.xlabel('Dane')
plt.ylabel('Wartość')
plt.show()

plt.figure(figsize=(8, 4))
plt.scatter(data['Dane'], data['Wartość'])
plt.title('Wykres kropkowy')
plt.xlabel('Dane')
plt.ylabel('Wartość')
plt.show()
```

```python
#zad 5:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


stacje = ['Stacja 1', 'Stacja 2', 'Stacja 3']
wysokosci = ['0 m', '500 m', '1000 m']
dni = ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-07']
temperatura = np.array([
    [[12, 15, 18], [15, 13, 9], [5, 7, 11]],
    [[13, 16, 19], [21, 22, 23], [9, 11, 13]],
    [[22, 25, 28], [12, 17, 21], [9, 10, 11]]
])
wilgotnosc = np.array([
    [[65, 75, 82], [56, 59, 71], [30, 45, 55]],
    [[44, 62, 63], [49, 61, 70], [38, 30, 32]],
    [[66, 67, 72], [62, 66, 71], [44, 55, 66]]
])


srednia_temperatura = np.mean(temperatura, axis=2)
srednia_wilgotnosc = np.mean(wilgotnosc, axis=2)

tablica_numpy_reshaped = tablica_numpy.reshape(3, 2, 3).swapaxes(1, 2).reshape(3, 6)

tablica_pandas = pd.DataFrame(tablica_numpy_reshaped, index=stacje, columns=pd.MultiIndex.from_product([wysokosci, ['Temperatura', 'Wilgotność']]))

print("Tablica wielowymiarowa (NumPy):")
print(tablica_numpy)
print("\nTablica wielowymiarowa (Pandas):")
print(tablica_pandas)


tablica_pandas.plot(kind='bar', figsize=(8, 4))
plt.title('Średnia temperatura i wilgotność na poszczególnych wysokościach')
plt.xlabel('Stacje')
plt.ylabel('Średnia')
plt.xticks(rotation=0)
plt.legend(['Temperatura', 'Wilgotność'])
plt.show()
```

```python
import numpy as np

# Tworzenie tablicy reprezentującej obraz
obraz = np.array([[30, 60, 130], [90, 120, 150], [175, 200, 255]])

min_wartosc = np.min(obraz)
max_wartosc = np.max(obraz)
srednia = np.mean(obraz)

print("Minimalna wartość pikseli:", min_wartosc)
print("Maksymalna wartość pikseli:", max_wartosc)
print("Średnia wartość pikseli:", srednia)
```

```python
import numpy as np

wyniki = np.array([[30, 80, 71], [25, 50, 90], [80, 75, 90], [65, 75, 90]])

udzial_zaliczonych = np.mean(wyniki >= 60, axis=0) * 100

zakresy_ocen = [0, 60, 70, 80, 101]
oceny = [2, 3, 4, 5]

warunki = [(wyniki >= zakresy_ocen[i]) & (wyniki < zakresy_ocen[i+1]) for i in range(len(zakresy_ocen) - 1)]
wyniki_ocen = np.select(warunki, oceny, default=oceny[-1])

srednie_oceny = np.mean(wyniki_ocen, axis=1)

print("Tablica wyników:\n", wyniki)
print("Procentowy udział zaliczonych testów:\n", udzial_zaliczonych)
print("Wyniki ocen:\n", wyniki_ocen)
print("Średnie oceny dla każdego studenta:\n", srednie_oceny)
```
